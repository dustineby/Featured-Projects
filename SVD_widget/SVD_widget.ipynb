```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
## mounting my google drive to grab a picture
## you can also add one manually

from google.colab import drive
drive.mount('/content/drive')
```

```python
## I use a picture of my cat, Flora, here
## but anyone using this will have to grab their own picture

## open the pic with matplotlib

plt.style.use('classic')
img = Image.open('/content/drive/MyDrive/Flora.bmp') # change to your pic location
img
```

```python
## convert pic to black and white

blackwhite = img.convert('LA')
blackwhite
```

```python
## use numpy to turn the picture into an array of greyscale values for each pixel
## and show the matrix

imgmat = np.array(list(blackwhite.getdata(band=0)),float)
imgmat.shape = (blackwhite.size[1], blackwhite.size[0])
imgmat
```

```python
## visualize the matrix to show that it does indeed contain the pictures information

plt.figure(figsize=(6,6))
plt.imshow(imgmat,cmap='gray')
plt.show()
```

```python
## use numpy to grab the 3 matrices from singular value decomposition on

V,D,W = np.linalg.svd(imgmat)
```

```python
## k = num of eiegenvectors to keep
## the higher this is, the more infromation we keep
k = 30

## show recreation after reducing the SVD matrices to size k (columns, diagonal entries, and rows respectively)
rec = np.matrix(V[:,:k]) * np.diag(D[:k]) * np.matrix(W[:k,:])
plt.figure(figsize=(6,6))
plt.imshow(rec, cmap = 'gray')
plt.show()
```
